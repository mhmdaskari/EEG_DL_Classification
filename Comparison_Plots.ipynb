## Plots


```python
import numpy as np
import matplotlib.pyplot as plt

"""
Creating the figure based on the runs of the best model, CNN,
and different inputs from 0 to 8 for the "sobject_id" input parameter of the
"data_train_valid_test_split" function in the "Best_Model_C247-Project-EEG-CNN.ipynb"
Notebook.
"""

data = {'1st':70.71, '2nd':50.50, '3rd':66.50,
        '4th':59.00,'5th':78.72, '6th':53.57, '7th':71.50,
        '8th':61.50, '9th':78.19}

courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (7, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='blue',
        width = 0.4)

plt.tick_params(axis='both', which='major', labelsize=13)

plt.ylabel("Accuracy (%)", fontsize=15)
plt.xlabel("Subjects", fontsize=15)
plt.title("Accuracy of Best Model Over Subjects", fontsize=15)
plt.ylim((0,100))
plt.yticks(np.arange(0, 101, step=10))
plt.axhline(y=70, c='red', linestyle='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
"""
Creating the figure based on the runs of the best model, CNN,
and different inputs from [100, 300, 500, 700, 1000] for the "max_timeStep" input parameter of the
"data_train_valid_test_split" and "CNN" functions in the "Best_Model_C247-Project-EEG-CNN.ipynb"
Notebook.
"""

data = {'100':53.61, '300':70.94, '500':70.49,
        '700':68.12,'1000':68.74}

courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (7, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='blue',
        width = 0.4)

plt.tick_params(axis='both', which='major', labelsize=13)
plt.xticks()
plt.ylabel("Accuracy (%)", fontsize=15)
plt.xlabel("Data Time Duration (ms)", fontsize=15)
plt.title("Accuracy of Best Model Over Time Durations", fontsize=15)
plt.ylim((0,100))
plt.yticks(np.arange(0, 101, step=10))
plt.axhline(y=70, c='red', linestyle='--')
plt.show()
```
